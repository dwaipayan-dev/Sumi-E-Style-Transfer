```python
for file in os.listdir(dir):
    img_path = os.path.join(dir, file)
img = cv2.imread(img_path)
img = cv2.resize(img, (224, 224))
mean = 0
var = 10
sigma = var ** 0.5
gaussian1 = np.random.normal(mean, sigma, (224, 224))
gaussian2 = np.random.normal(mean, sigma, (224, 224))
gaussian3 = np.random.normal(mean, sigma, (224, 224))
noisy_image = np.zeros(img.shape, np.float32)

if len(img.shape) == 2:
    noisy_image = img * gaussian
else:
    noisy_image[:, :, 0] = img[:, :, 0] * gaussian1
    noisy_image[:, :, 1] = img[:, :, 1] * gaussian2
    noisy_image[:, :, 2] = img[:, :, 2] * gaussian3
noise2 = (noisy_image - noisy_image.min())/(noisy_image.max() - noisy_image.min())

plt.imshow(noise2)

```
